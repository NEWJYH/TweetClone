# Home.js
## Form 작업

```jsx
import { useState } from "react";

function Home() {
  const [jweet, setJweet] = useState("");

  const onSubmit = (event) => {
    event.preventDefault();
    if (jweet) {
      console.log(jweet);
      // 서버 전송 로직
      setJweet("");
    }
  };
  const onChange = (event) => {
    const {
      target: { value },
    } = event;
    setJweet(value);
  };

  return (
    <>
      <form onSubmit={onSubmit}>
        <input
          type="text"
          placeholder="What's on your mind?"
          maxLength={120}
          onChange={onChange}
          value={jweet}
        />
        <input type="submit" value="Jweet" />
      </form>
    </>
  );
}

export default Home;

```
